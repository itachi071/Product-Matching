```python
import pandas as pd
import numpy as np
```

```python
amazon=pd.read_csv('amz_com-ecommerce_sample.csv',encoding= 'unicode_escape')
fipkart=pd.read_csv('flipkart_com-ecommerce_sample.csv',encoding= 'unicode_escape')
```

```python
amazon.head()
```

```python
fipkart.head()
```

```python
a=amazon[["product_name","retail_price","discounted_price"]]
f=fipkart[["product_name","retail_price","discounted_price"]]
```

```python
a
```

```python
f
```

```python
a['product_name']=a['product_name'].apply(lambda x:x.lower())
f['product_name']=f['product_name'].apply(lambda x:x.lower())
```

```python
a['fipkart_pro_name']=f['product_name']
```

```python
a['com'] = np.where(a['product_name'] == a['fipkart_pro_name'], a['product_name'], np.nan)
```

```python
a[a['com'].isnull()==True]
```

```python
a['f_retail_price']=f['retail_price']
a['f_discounted_price']=f['discounted_price']
```

```python
a.isnull().sum()
```

```python
a.dropna(subset=['com'],inplace=True)
```

```python
a.reset_index(inplace=True,drop=True)
```

```python
a.drop(['com'],axis=1, inplace=True)
```

```python
a
```

```python
a[a['f_retail_price'].isnull()==True]
```

```python
a.dropna(subset=['f_retail_price'],inplace=True)
```

```python
a
```

```python
a.to_csv('Task_2.csv')
```
